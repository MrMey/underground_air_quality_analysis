#### Date & Time processing

```python
data['year'],data['month'],data['day'] = data["date"].str.split("-").str
data['year'] = data['year'].astype(int)
data['month'] = data['month'].astype(int)
data['day'] = data['day'].astype(int)
```

```python
data['hour'],data['minute'],data['second'] = data["time"].str.split(":").str
data['hour'] = data['hour'].astype(int)
data['minute'] = data['minute'].astype(int)
data['second'] = data['second'].astype(int)
```

```python
data['date'] = pd.to_datetime(data['date'], format = "%Y-%m-%d")
```

```python
data['time'] = pd.to_datetime(data['time'], format = "%H:%M:%S")
```

```python
data['time'] = data["time"].dt.time
```

```python
print(len(data[data["co2"].isnull() & 
                        data["humi"].isnull() &
                        data["no"].isnull() & 
                        data["no2"].isnull() &
                        data["pm10"].isnull() & 
                        data["pm2.5"].isnull() & 
                        data["no2"].isnull() & 
                       data["temp"].isnull()]))
```

For 15860 raws, data is missing for almost all sensors (except rain) -> remove all these rows

```python
missing_data = data[data["co2"].isnull() | 
                        data["humi"].isnull() |
                        data["no"].isnull() | 
                        data["no2"].isnull() |
                        data["pm10"].isnull() | 
                        data["pm2.5"].isnull() | 
                        data["no2"].isnull() |
                       data["temp"].isnull()]
```

```python
data = data.dropna()
```

```python
ax = sns.factorplot(x="year", y="temp", hue = "station", data=data, kind = "box")
```

```python
data_2013 = data[data['year'] == "2013"]
```

```python
fig, ax = plt.subplots(figsize=(20,10))
data_2013.plot(x = 'date',y = 'co2', ax = ax)
```

```python
daily_avg_data = data.groupby(["date"]).mean().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
daily_avg_data[daily_avg_data["station"] == "auber"].plot(x = "date",y = "co2", ax = ax, label = "auber")
```

```python
weekly_avg_data = data.groupby([pd.Grouper(key = 'date',freq = 'W')]).mean().reset_index()
```
