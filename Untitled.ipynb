## Feature engineering

```python
import numpy as np
```

```python
data['co2_log'] = data['co2'].apply(lambda x:np.log(x))
data['precipitation_log'] = data['precipitation'].apply(lambda x:np.log(x))
```

## Modelization

```python
features = ['frequentation','year','month','hour', 'weekday', 'precipitation','temp_min','temp_max',]
X = data[features].values
Y = data['co2'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn import ensemble
```

```python
cv_ratio = 0.8
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

#### Ridge Regression

```python
clf = linear_model.Lasso(alpha = 0.1)
clf.fit(X_train,Y_train)
clf.score(X_test,Y_test)
```

```python
for idx in range(len(features)):
    print(features[idx],clf.coef_[idx])
```

#### ElasticNet Regression

```python
clf = linear_model.ElasticNet(alpha = 0.1)
clf.fit(X_train,Y_train)
clf.score(X_test,Y_test)
```

```python
for idx in range(len(features)):
    print(features[idx],clf.coef_[idx])
```

#### Random Forest Regression

```python
clf = ensemble.RandomForestRegressor(n_estimators = 100)
clf.fit(X_train,Y_train)
clf.score(X_test,Y_test)
```

```python
for idx in range(len(features)):
    print(features[idx],clf.feature_importances_[idx])
```
